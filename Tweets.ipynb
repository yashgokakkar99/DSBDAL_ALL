```python
import pandas as pd
import numpy as np
from textblob import TextBlob

df = pd.read_csv('tweets.csv')
```

```python
df['tweet'] = df['tweet'].str.lower()
df['tweet'] = df['tweet'].replace('rt', '')
df['tweet'] = df['tweet'].replace('@', '')
df['tweet'] = df['tweet'].replace('#', '')
```

```python
df['sentiment'] = df['tweet'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['sentiment'] = df['sentiment'].apply(lambda x: 'positive' if x > 0 else 'negative' if x < 0 else 'neutral')
```
